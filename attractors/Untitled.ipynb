```python
import numpy as np
from numpy import sin, cos
import matplotlib.pyplot as plt
```

```python
p = 10000
x = 0
y = 0
x1 = 0
y1 = 0
# a = -1.4
# b = 1.6
# c = 1.0
# d = 0.7

a = -1.7
b= 1.8
c = -1.9
d = -0.4   

xs = np.zeros(p)
ys = np.zeros(p)

for i in range(p):
    xn = sin(a*y1)+(c*cos(a*x1))
    yn = sin(b*x1)+(d*cos(a*y1))
    
    x1, y1 = xn, yn
    xs[i] = x1
    ys[i] = y1
    
```

```python
min(xs)
```

```python
fig, ax = plt.subplots()
fig.set_size_inches((10, 10))
fig.patch.set_visible(True)
fig.patch.set_facecolor('black')
ax.axis('off')


plt.plot(xs, ys)

# plt.scatter(xs, ys, c=ys, cmap='twilight', marker='1')
plt.show()
```
