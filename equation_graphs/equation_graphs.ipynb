```python
def f2(x):
    if x < 0:
        firstpart = (abs(x)**(2/3))
    else:
        firstpart = x**(2/3)
    
    xsquared = x**2
    inner = 3.3-xsquared
    outer = 0.9*np.sqrt(inner)
    
    sininner = 15*np.pi*x
    sinval = np.sin(sininner)

    val = firstpart + (outer * sinval)
    # print(val)
    return val
```
