```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import numpy as np
```

```python
# background = 'aliceblue'
background = '#042b1a'

#define Matplotlib figure and axis
fig, ax = plt.subplots()

fig.set_size_inches((20,20))

#create simple line plot
ax.plot([1, 1],[2, 2])

#add rectangle to plot
ax.add_patch(Rectangle((10, 1), 6, 6, color='red', angle=0))
ax.add_patch(Rectangle((2, 7), 8, 8, color='maroon', angle=0))
ax.add_patch(Rectangle((10, 15), 10, 10, color='coral', angle=-53))

plt.gca().set_aspect('equal', adjustable='box')
# plt.gca().set_aspect('auto', adjustable='box')
fig.patch.set_facecolor(background)
ax.axis('off')

#display plot
plt.savefig('temp.png', bbox_inches='tight', pad_inches=0, dpi=400)
plt.show()
```
