```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 100 linearly spaced numbers
x = np.linspace(-20,20,100)

# the function, which is y = x^2 here
y = abs((6*x)+1)

fig, ax = plt.subplots()
fig.set_size_inches(7, 5)
fig.patch.set_visible(True)
fig.patch.set_facecolor('white')
ax.axis('off')


# plot the function
plt.plot(x,y, 'r')
x = np.linspace(-20,200000, 200000)
tps = \
[(3,5), (5,7), (11,13), (17,19), (29,31), (41,43), (59,61), (71,73), (101, 103), (107, 109), 
 (137, 139), (149, 151), (179, 181), (191, 193), (197, 199), (227, 229), (239, 241), (269, 271), (281, 283), (311, 313),
 (347, 349), (419, 421), (431, 433), (461, 463), (521, 523), (569, 571), (599, 601), (617, 619), (641, 643), (659, 661),
 (809, 811), (821, 823), (827, 829), (857, 859), (881, 883), (1019, 1021)]
colors = plt.cm.summer(np.linspace(0,1,len(tps)))

for i, t in enumerate(tps):
    m = (t[0]+1)/6
    y = (x/m)+(6*m)
    plt.plot(x,y, color=colors[i])
    
# show the plot
plt.xlim([-5, 200000])
plt.ylim([0, 3000])
# plt.savefig('image.png', bbox_inches='tight',pad_inches = 0)
plt.savefig('images/white_summer.png', bbox_inches='tight', pad_inches=0)
plt.show()
```
