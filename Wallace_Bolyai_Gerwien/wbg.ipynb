```python
import numpy as np 
import matplotlib.pyplot as plt

x1 = 4.33*np.cos(np.deg2rad(30))
y1 = 4.33*np.sin(np.deg2rad(30))

fig, ax = plt.subplots()
fig.set_size_inches((20,20))

# blue
t1 = plt.Polygon([(0,0),(5,0),(x1,y1)])
plt.gca().add_patch(t1)

# red
t2 = plt.Polygon([(0,0),(0,np.sqrt(75)),(x1,y1)], color='r')
plt.gca().add_patch(t2)

# green
xx = 0 - (2 * np.cos(np.deg2rad(60)))
yy = np.sqrt(75) - (2 * np.sin(np.deg2rad(60)))
t3 = plt.Polygon([(0,(np.sqrt(75)-3)),(0,np.sqrt(75)),(xx,yy)], color='g')
plt.gca().add_patch(t3)

# magenta
xx2 = xx + (1 * np.cos(np.deg2rad(307)))
yy2 = yy + (1 * np.sin(np.deg2rad(307)))
t4 = plt.Polygon([(-5,0),(xx,yy),(xx2,yy2)], color='m')
plt.gca().add_patch(t4)


# cyan
t5 = plt.Polygon([(-5,0),(xx2,yy2),(0,np.sqrt(75)-3),(0,0)], color='c')
plt.gca().add_patch(t5)


# t2 = plt.Polygon([(0,temp),(0,np.sqrt(75)),(-xx,yy)], color='g')
# plt.gca().add_patch(t2)

#blue

# y1 = -(np.sqrt(75)/5)*3+np.sqrt(75)
# t1 = plt.Polygon([(0,0),(3,y1),(0,np.sqrt(75))])
# plt.gca().add_patch(t1)

# #red
# t2 = plt.Polygon([(0,0),(5,0),(3,y1)], color='r')
# plt.gca().add_patch(t2)

# #green
# y2 = (np.sqrt(75)/5)*-1.25+np.sqrt(75)
# t3 = plt.Polygon([(0,6),(0,np.sqrt(75)),(-1.25,y2)], color='g')
# plt.gca().add_patch(t3)

plt.ylim(-10,10)
plt.xlim(-10,10)

plt.show()
```
