```python
from mpmath import mp
import matplotlib.pyplot as plt
import numpy as np
```

```python
def transforms(x):
    seq=[0]
    val=[0]
    rad=0
    
    for i in range(1, len(x)):
        even= (190*x[i])* (np.pi / 180 )
        rad=rad+even
        seq.append(seq[i-1]+np.sin(rad))
        # odd = 60* (np.pi / 180 )
        # if x[i]%2==0:
        #     seq.append(seq[i-1]+np.sin(rad+even))
        #     rad=rad+even            
        # else:
        #     seq.append(seq[i-1]+np.sin(rad+odd))
        #     rad=rad+odd
        val.append(val[i-1]+np.cos(rad))
    return val,seq
```

```python
fig, ax = plt.subplots()
fig.set_size_inches((20,10))
fig.patch.set_visible(True)
fig.patch.set_facecolor('black')
ax.grid(False)

mp.dps = 20000  # set number of digits
# number = mp.e
number = mp.pi
# number = mp.phi
number = mp.euler

num_str = str(number)
num_str = num_str.replace('.', '')

int_n = [int(n) for n in num_str]

# trace = transforms(int_n)

x,y = transforms(int_n)
#ax.set_facecolor('black')
ax.plot(x,y, c='c', linewidth=1.0)#color=color_picker());  

ax.axis('off')
plt.show()

```
